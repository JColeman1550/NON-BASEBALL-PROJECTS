```python
%%capture
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure #Adjusts  configurations of plots 

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#read in data
path = "/content/drive/MyDrive/movies.csv"
```

```python
#read in data
movie_data = pd.read_csv(path)
```

```python
movie_data.head()
```

```python
# Cleaning Data (Findidng Missing data, Find Null)

for col in movie_data.columns:
  pct_missing = np.mean(movie_data[col].isnull())
  print('{} - {}%'.format(col, pct_missing))

movie_data.dropna()
```

```python
# Find Data Types

movie_data.dtypes
```

```python
#create correct year column bc after scanning entire dataset some of the  years did not matchup with released column

movie_data['year_correct'] = movie_data['released'].astype(str).str[:13]
movie_data.head()
```

```python
#order data by gross revenue
movie_data.sort_values(by=['gross'], inplace=True, ascending=False)
movie_data.head()
```

```python
#Viewing entire spreadsheet as default
pd.set_option('display.max_rows',None)
```

```python
#Dropping duplicates
movie_data['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#What correlates to gross revenue? Budget? Comapny?

#Building scatterplot Budget v Gross

plt.scatter( x=movie_data['budget'], y=movie_data['gross'])

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget for Film')

plt.ylabel('Gross Earnings')
plt.show()
```

```python
movie_data.head()
```

```python
#regression plot budget v gross (seaborn)

sns.regplot(x='budget',y='gross', data=movie_data, scatter_kws={'color':'red'}, line_kws={"color":"blue"})
```

```python
# Analyzing Correlation
#only works on numerical fields
#Types of correlation -Pearson(default), Kendall, spearman
movie_data.corr(method = 'pearson')
```

```python
#high correlation between budget and gross exist also gross and votes
```

```python
correlation_matrix = movie_data.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
```

```python
correlation_matrix = movie_data.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.scatter( x=movie_data['budget'], y=movie_data['gross'])

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')
plt.show()
```

```python
#Company has  correlation? Making company names a numeric value

movie_data.head()
```

```python
movie_data_numerized = movie_data

for col_name in movie_data_numerized:
  if(movie_data_numerized[col_name].dtype == 'object'):
    movie_data_numerized[col_name] = movie_data_numerized[col_name].astype('category')
    movie_data_numerized[col_name] = movie_data_numerized[col_name].cat.codes


movie_data_numerized.head()
```

```python
correlation_matrix = movie_data.corr(method = 'pearson')

sns.heatmap(correlation_matrix, annot=True)
plt.scatter( x=movie_data['budget'], y=movie_data['gross'])

plt.title('Correlation Matrix For Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')
plt.show()
```

```python
movie_data_numerized.corr()
```

```python
#unstacking
correlation_mat = movie_data_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
